```python
#Dependancies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import random
import matplotlib.animation
import matplotlib.patches as mpatches

from scipy.integrate import solve_ivp
from scipy.interpolate import CubicSpline
from scipy.optimize import fsolve, least_squares, minimize
from scipy.stats import qmc 
from scipy.linalg import norm
from matplotlib.collections import LineCollection
from matplotlib.colors import ListedColormap, BoundaryNorm
from scipy.signal import argrelmin, argrelmax
from numba import jit
from tabulate import tabulate
from itertools import product
from operator import itemgetter
```

### System model - with Pu damping and adsorption cut-off

```python
@jit
def model(t, X, V_max, γ_Z, φ, g, ν_x, λ_P, λ_Z, λ_Z_hat, λ_E, δ, μ_V, μ_V_prime, 
          μ_u, μ_r, μ_s, μ_P, μ_delta, μ_g, μ_Z, K_N, K_I, K_h, K_P, I_0, ω):
    
    """The NPZV system, formulated with 7 state variables defined below. 
    Defined to be used with the numerical integrating function, solve_ivp."""
    
    N_n = X[0] #New nutrients (Nitrate)
    N_r = X[1] #Recycled Nutrients (Ammonium)
    P_U = X[2] #Susceptible Hosts
    P_I = X[3] #Infected Hosts
    Z   = X[4] #Zooplankton
    V_I = X[5] #Internal viruses from host nucleotide recycling 
               #and de novo nucleotide synthesis
    V_E = X[6] #Free (extracellular) viruses 
    
    alpha = (1 / 1.42e8) * 2.1e-10 * 1e6 # host conversion, mmol ml / NT m^3
    beta  = (1 / 4e5) * 1.27e-15 * 1e6   # viral conversion, mmol ml / NT m^3
    
    if V_I == 0:  ν_i = 0 #avoid divison by 0 when evaluating ν_i
    else: ν_i = V_I / ((beta / alpha) * P_I + V_I)
     
    #Useful abreviations
    P = P_U + P_I
    N = N_n + N_r
    nutr_lim = N / (K_N + N)

    light_lim = np.log((K_I + I_0 / K_I)) / K_h  #light limitation term
    
    psi = V_max * nutr_lim * light_lim
    
    #Defining the scaling factor that continuously reduces the affect of growth
    #as P_U reaches ecologically unrealistic densities (e.g. 1e-20)
    thresh = 1e-12 #threshold for when growth damping begins
    P_U_scale = 1
    
    if P_U <= thresh: P_U_scale = P_U / thresh 

    #V_E minimum for adsportion cut off
    V_E_min = 1e-12
    
    #Uninfected Phytoplankton
    P_U_growth  = P_U_scale * (P_U + μ_u * P_I) * psi
    
    P_U_grazing = P_U * g * Z / K_P
    P_U_mort    = P_U * λ_P
    P_U_adsorp  = P_U * (1 - ν_x) * φ * μ_s * max(0, V_E - V_E_min)
    P_U_entrain = P_U * ω 
    
    P_U_dt = P_U_growth - P_U_grazing - P_U_mort - P_U_adsorp \
             - P_U_entrain 
    
    #Infected Phytoplankton
    P_I_growth   = (1 - P_U_scale * μ_u) * P_I * psi
    P_I_adsorp   = (1 - ν_x) * φ * μ_s * P_U * max(0, V_E - V_E_min)
    
    P_I_grazing  = P_I * g * Z / K_P
    P_I_mort     = P_I * λ_P
    P_I_nt_recyc = P_I * ν_i * μ_V
    P_I_lysis    = P_I * ν_i * δ
    P_I_entrain  = P_I * ω
    
    P_I_dt = P_I_growth + P_I_adsorp - P_I_grazing \
            - P_I_mort - P_I_nt_recyc - P_I_lysis - P_I_entrain
    
    #Zooplankton
    Z_assim   = Z * γ_Z * g * (P + V_I) / K_P
    Z_mort    = Z * (λ_Z + λ_Z_hat * Z)
    Z_entrain = Z * ω

    Z_dt = Z_assim - Z_mort - Z_entrain
    
    #Intracellular viruses
    V_I_nt_recyc   = (beta / alpha) * ν_i * μ_V * P_I
    V_I_denovo_syn = (beta / alpha) * ν_i * nutr_lim * μ_V_prime * P_I
    
    V_I_growth  = V_I_nt_recyc + V_I_denovo_syn
    V_I_adsorp  = (beta / alpha) * ν_x * φ * μ_s * P_U * max(0, V_E - V_E_min)
    
    V_I_grazing = V_I * g * Z / K_P
    V_I_lysis   = V_I * ν_i * δ
    V_I_h_mort  = V_I * λ_P
    V_I_entrain = V_I * ω
    
    V_I_dt = V_I_growth + V_I_adsorp - V_I_lysis \
            - V_I_grazing - V_I_h_mort - V_I_entrain
    
    #Extracellular viruses
    V_E_h_mort  = V_I * μ_r * λ_P

    V_E_adsorp  = (beta / alpha) * ν_x * φ * μ_s * P_U * max(0, V_E - V_E_min)
    V_E_mort    = λ_E * V_E
    V_E_entrain = V_E * ω
    
    V_E_dt = V_I_lysis + V_E_h_mort - V_E_adsorp - V_E_mort - V_E_entrain
    
    # Aviod division by zero
    if N == 0:
        Nn_partition, Nr_partition = 0, 0
    else: 
        Nn_partition, Nr_partition = (N_n / N), (N_r / N)
        
    #Nitrate
    N_n_deplet    = Nn_partition * (P_U_growth + P_I_growth) #scaling terms in previous def
    N_n_intracell = Nn_partition * V_I_denovo_syn
    N_n_entrain   = ω * N_n 
    
    N_n_dt = - N_n_deplet - N_n_intracell - N_n_entrain
            
    #Recycled nutrients
    N_r_deplet    = Nr_partition * (P_U_growth + P_I_growth) 
    
    N_r_nt_recyc   = ν_i * P_I * (1 - beta / alpha) * μ_V
    N_r_denovo_syn = Nr_partition * V_I_denovo_syn
    
    N_r_intracell  = N_r_nt_recyc - N_r_denovo_syn
    
    N_r_mort    = λ_P * (P + (1 - μ_r) * V_I)
    N_r_sloppy  = (1 - γ_Z) * g * Z * (P + V_I) / K_P
    N_r_entrain = N_r * ω
    
    N_r_remin   = (μ_P * N_r_mort) + V_E_mort + (μ_delta * P_I_lysis) \
                + (μ_g * N_r_sloppy) + (μ_Z * Z_mort) - N_r_entrain

    N_r_export  = (1 - μ_P) * N_r_mort + (1 - μ_delta) * P_I_lysis \
                + (1 - μ_g) * N_r_sloppy + (1 - μ_Z) * Z_mort
    
    N_r_dt = - N_r_deplet + N_r_intracell + N_r_remin + N_r_export
            
    
    return N_n_dt, N_r_dt, P_U_dt, P_I_dt, Z_dt, V_I_dt, V_E_dt
```

### Sources vs. Sinks Data Generator

```python
def svs(sol, param):
    
    """Sources and sinks of each constituant in the coupled NPZV model.
    
    sol: solve_ivp solution object
    param: all parameters needed for calulation
    
    return: data set for each state with individual sources and sinks
    """
    
    # Parameter definitions (floats)
    V_max, γ_Z, φ, g, ν_x, λ_P, λ_Z, λ_Z_hat, λ_E, δ, μ_V, μ_V_prime, \
    μ_u, μ_r, μ_s, μ_P, μ_delta, μ_g, μ_Z, K_N, K_I, K_h, K_P, I_0, ω = param
    
    L = len(sol.t)
    
    # Defineing state solutions (arrays)
    N_n = sol.y[0] # New nutrients (Nitrate)
    N_r = sol.y[1] # Recycled Nutrients (Ammonium)
    P_U = sol.y[2] # Susceptible Hosts
    P_I = sol.y[3] # Infected Hosts
    Z   = sol.y[4] # Zooplankton
    V_I = sol.y[5] # Internal viruses from host NT recycling and de novo NT synthesis
    V_E = sol.y[6] # Free (extracellular) viruses 
    
    alpha = (1 / 1.42e8) * 2.1e-10 * 1e6 # host conversion, mmol ml / NT m^3
    beta  = (1 / 4e5) * 1.27e-15 * 1e6   # viral conversion, mmol ml / NT m^3
    
    # Useful abriviations
    P = P_U + P_I                                #array
    N = N_n + N_r                                #array
    nutr_lim = N / (K_N + N)                     #array
    light_lim = np.log((K_I + I_0 / K_I)) / K_h  #float
    psi = V_max * nutr_lim * light_lim           #array
    
    ν_i = np.zeros(L)
    P_U_scale = [] # scale only applies Pu
    
    # Defing arrays that require specific logic at each step
    for i in range(L):
        
        if N[i] == 0: N[i] = 1 # avoid divison by 0 when evaluating nutrient ratios
                               # *** Does NOT affect nutrient limitation terms

        if V_I[i] != 0: ν_i[i] = V_I[i] / ((beta / alpha) * P_I[i] + V_I[i])

        # Defining damping operator
        scale = 1
        thresh = 1e-12
        if P_U[i] < thresh: scale = P_U[i] / thresh
        P_U_scale.append(scale)
    
    P_U_scale = np.array(P_U_scale)
    adsorp_sub_VE = V_E - np.full(L, 1e-12) 
                      
    for i in range(L):
        # creation of adsorption cut off - same as max(0, Ve-Ve_min)
        if adsorp_sub_VE[i] < 0:
            adsorp_sub_VE[i] = 0
    
    # Uninfected Phytoplankton
    P_U_svs = {"Time (Days)": sol.t, "P_U": P_U}
    
    P_U_svs["Growth"]    = P_U_scale * (P_U + μ_u * P_I) * psi
    
    P_U_svs["Grazing --"]   = P_U * g * Z / K_P
    P_U_svs["Mortality --"] = P_U * λ_P
    P_U_svs["Adsorption --"]    = P_U * (1 - ν_x) * φ * μ_s * adsorp_sub_VE
    P_U_svs["Entrainment --"]   = P_U * ω 
    
    P_U_svs["Sources"] = P_U_svs["Growth"]
    P_U_svs["Sinks"]   = P_U_svs["Grazing --"] + P_U_svs["Mortality --"] \
                         + P_U_svs["Adsorption --"] + P_U_svs["Entrainment --"]
    
    P_U_svs["dPudt"]  = P_U_svs["Sources"] - P_U_svs["Sinks"]
    
    
    # Infected Phytoplankton transfer rates
    P_I_svs = {"Time (Days)": sol.t, "P_I": P_I}
    
    P_I_svs["Growth"]    = (1 - P_U_scale * μ_u) * P_I * psi
    P_I_svs["Adsorption"]    = (1 - ν_x) * φ * μ_s * P_U * adsorp_sub_VE
    
    P_I_svs["Grazing --"]   = P_I * g * Z / K_P
    P_I_svs["Mortality --"] = P_I * λ_P
    P_I_svs["NT Recycling --"]   = P_I * ν_i * μ_V
    P_I_svs["Lysis --"]     = P_I * ν_i * δ
    P_I_svs["Entrainment --"]   = P_I * ω
    
    P_I_svs["Sources"] = P_I_svs["Growth"] + P_I_svs["Adsorption"]
    P_I_svs["Sinks"]   = P_I_svs["Grazing --"] + P_I_svs["Mortality --"] \
                         + P_I_svs["NT Recycling --"] + P_I_svs["Lysis --"] \
                         + P_I_svs["Entrainment --"]
    
    P_I_svs["dPidt"]  = P_I_svs["Sources"] - P_I_svs["Sinks"]
    
    
    # Zooplankton
    Z_svs = {"Time (Days)": sol.t, "Z": Z}
    
    Z_svs["Assimilation"] = Z * γ_Z * g * (P + V_I) / K_P
    
    Z_svs["Mortality --"]    = Z * (λ_Z + λ_Z_hat * Z)
    Z_svs["Entrainment --"]      = Z * ω

    Z_svs["Sources"] = Z_svs["Assimilation"] 
    Z_svs["Sinks"]   = Z_svs["Mortality --"] + Z_svs["Entrainment --"]
    
    Z_svs["dZdt"] = Z_svs["Sources"] - Z_svs["Sinks"]
    
    
    # Intracellular viruses
    V_I_svs = {"Time (Days)": sol.t, "V_I": V_I}
    
    V_I_nt_recyc  = (beta / alpha) * ν_i * μ_V * P_I
    V_I_denovo_syn = (beta / alpha) * ν_i * nutr_lim * μ_V_prime * P_I
    
    V_I_svs["Growth"] = V_I_nt_recyc + V_I_denovo_syn
    V_I_svs["Adsorption"] = (beta / alpha) * ν_x * φ * μ_s * P_U * adsorp_sub_VE
    
    V_I_svs["Grazing --"]        = V_I * g * Z / K_P
    V_I_svs["Lysis --"]          = V_I * ν_i * δ
    V_I_svs["Host Mortality --"] = V_I * λ_P
    V_I_svs["Entrainment --"]    = V_I * ω
    
    V_I_svs["Sources"] = V_I_svs["Growth"] + V_I_svs["Adsorption"]
    V_I_svs["Sinks"]   = V_I_svs["Grazing --"] + V_I_svs["Lysis --"] \
                         + V_I_svs["Host Mortality --"] + V_I_svs["Entrainment --"]
    
    V_I_svs["dVidt"] = V_I_svs["Sources"] - V_I_svs["Sinks"]
    
    
    # Extracellular viruses
    V_E_svs = {"Time (Days)": sol.t, "V_E": V_E}
    
    V_E_svs["Host Mortality"] = V_I * μ_r * λ_P
    V_E_svs["Lysis"] = V_I_svs["Lysis --"]
    
    V_E_svs["Adsorption --"]  = (beta / alpha) * ν_x * φ * μ_s * P_U * adsorp_sub_VE
    V_E_svs["Mortality --"]   = λ_E * V_E
    V_E_svs["Entrainment --"] = V_E * ω
    
    V_E_svs["Sources"] = V_E_svs["Lysis"] + V_E_svs["Host Mortality"]
    V_E_svs["Sinks"]   = V_E_svs["Adsorption --"] + V_E_svs["Mortality --"] \
                         + V_E_svs["Entrainment --"]
    
    V_E_svs["dVEdt"] = V_E_svs["Sources"] - V_E_svs["Sinks"]
    
    
    # Nitrate
    N_n_svs = {"Time (Days)": sol.t, "N_n": N_n}
    
    N_n_svs["Depletion --"]              = (N_n / N) * (P_U_svs["Growth"] + P_I_svs["Growth"]) #scaling terms in previous def
    N_n_svs["Inracellular Processes --"] = (N_n / N) * V_I_denovo_syn
    N_n_svs["Entrainment --"]            = N_n * ω
    
    N_n_svs["Sources"] = np.zeros(L)
    N_n_svs["Sinks"] = N_n_svs["Depletion --"] + N_n_svs["Inracellular Processes --"] \
                        + N_n_svs["Entrainment --"]
    
    N_n_svs["dNndt"] = - N_n_svs["Sinks"]
    
    
    # Recycled nutrients
    N_r_svs = {"Time (Days)": sol.t, "N_r": N_r}
    
    N_r_svs["Depletion --"] = (N_r / N) * (P_U_svs["Growth"] + P_I_svs["Growth"])
    
    N_r_nt_recyc  = ν_i * P_I * (1 - beta / alpha) * μ_V
    N_r_denovo_syn = (N_r / N) * V_I_denovo_syn
    
    N_r_svs["Intracellular Processes"] = N_r_nt_recyc - N_r_denovo_syn
    
    N_r_svs["Mortality"]     = λ_P * (P + (1 - μ_r) * V_I)
    N_r_svs["Sloppy Eating"] = (1 - γ_Z) * g * Z * (P + V_I) / K_P
    N_r_svs["Entrainment --"]   = N_r * ω
    
    N_r_svs["Remineralization"] = (μ_P * N_r_svs["Mortality"]) + V_E_svs["Mortality --"] \
                                   + (μ_delta * P_I_svs["Lysis --"]) + (μ_g * N_r_svs["Sloppy Eating"]) \
                                   + (μ_Z * Z_svs["Mortality --"]) - N_r_svs["Entrainment --"]

    N_r_svs["Export"] = (1 - μ_P) * N_r_svs["Mortality"] + (1 - μ_delta) * P_I_svs["Lysis --"] \
                + (1 - μ_g) * N_r_svs["Sloppy Eating"] + (1 - μ_Z) * Z_svs["Mortality --"]
    
    N_r_svs["Sources"] = N_r_svs["Intracellular Processes"] + N_r_svs["Remineralization"] + N_r_svs["Export"]
    N_r_svs["Sinks"] = N_r_svs["Depletion --"]
    
    N_r_svs["dNrdt"] = N_r_svs["Sources"] - N_r_svs["Sinks"]
    
    P_U_svs = pd.DataFrame.from_dict(P_U_svs)
    P_I_svs = pd.DataFrame.from_dict(P_I_svs)
    Z_svs   = pd.DataFrame.from_dict(Z_svs)
    V_I_svs = pd.DataFrame.from_dict(V_I_svs)
    V_E_svs = pd.DataFrame.from_dict(V_E_svs)
    N_n_svs = pd.DataFrame.from_dict(N_n_svs)
    N_r_svs = pd.DataFrame.from_dict(N_r_svs)
    
    return N_n_svs, N_r_svs, P_U_svs, P_I_svs, Z_svs, V_I_svs, V_E_svs
```

### Perturb given initial condition arrays

```python
def perturb_range(ranges, range_names, perturb=True, pprint=False):
    
    """Perturb each member of a list.
    Doesn't perturn last value..."""
    
    perturbed_ranges = []
    
    if perturb:
        
        for state_range in ranges:
            temp_list = state_range.copy()

            for i in range(0, len(temp_list) - 1):
                range_gap = (temp_list[i + 1] - temp_list[i])
                pert = random.uniform(0, range_gap)

                temp_list[i] += pert

            #To increase reproducability, round ranges to 4 decimal places
            temp_list = np.round(temp_list, 4)
            perturbed_ranges.append(temp_list)
            
    else: 
        for state_range in ranges:
            perturbed_ranges.append(np.round(state_range, 4))

    if pprint:
        fig, axs = plt.subplots(1, 4, figsize = (12, 3))
        
        for r, state_range in enumerate(ranges):
            
            axs[r].plot(state_range, 'o', color = 'blue', label = 'Original', alpha=.5)
            axs[r].plot(perturbed_ranges[r], 'o', color = 'red', label="Perturbed")

            axs[r].set_title(range_names[r] + " Range")
            axs[r].set_xlabel("Counter")
            axs[r].set_ylabel("Parameter Value")
            axs[r].set_yscale("log")
            axs[r].legend()
            axs[r].grid(alpha=.3)

        plt.tight_layout()
        plt.show()
        
    return perturbed_ranges
```

### Fancy print parameterization

```python
def print_parameterization(param):
    
    """Print model parameters in a fancy table."""
    
    param_labels_fp = ['V_max', 'γ_Z', 'φ', 'g', 'ν_x = V_ind / (V_ind + S_ind)', 
                   'λ_P', 'λ_Z', 'λ_Z_hat', 'λ_E', 'δ', 'μ_V', 'μ_V_prime', 'μ_u', 'μ_r', 'μ_s', 
                   'μ_P', 'μ_delta', 'μ_g', 'μ_Z', 'K_N', 'K_I', 'K_h', 'K_P', 'I_0 = e - 1', 'ω']

    table = np.empty((5,5), dtype=object)
    i = 0

    for col in range(5):
        for row in range(5):

            table[row][col] = f"{param_labels_fp[i]} = {round(param[i], 4)}"
            i += 1

    print(tabulate(table, tablefmt='fancy_grid'))
    
    return
```

### Scheme for determining end state behavior

```python
def ss_or_osc(sol):
    """
    sol: is solution object resulting from the solve_ivp integration
    state_sol: an individual soltuion for a particular state variable 
    associated with the larger solution set, sol.
    
    Convergence Criteria:

    - For state solution state_sol within the solution, set reference value to the 
      last time iterate of the state, state_sol[-1].
      
    - Work backwards, comparing each time iterate to the reference value: want to find 
      the first instance of when the state_sol has strayed too far from the reference,
      as determined by a specified threshold.

    - That first instance represents the time of convergence.
    - Example: l = 5 (solution with 5 time steps)
               ref = sol[-1] = sol[4]
               convergence at i = 3 yields sol[-4] = sol[2] = sol(5 - 3)

    """
    ss = []        
    conv_index = [] 

    for k in range(7): #looping through state variables
        
        state_sol = sol.y[k]

        ### Check if oscillations occur:
        chunk = int(len(sol.t) * .70) #Last 30% of time span
        y_end = state_sol[chunk: ]    #look at tail end of specific state
        
        l_min = argrelmin(y_end)[-1]  #Extracting local minimum 
        l_max = argrelmax(y_end)[-1]  #and maximum
        
        if (len(l_min) != 0) and (len(l_max) != 0): 
            
            ref = (y_end[l_max[-1]] + y_end[l_min[-1]]) / 2
            
            if ref > 1e-12:
                
                wave_height = y_end[l_max[-1]] - y_end[l_min[-1]]
                amplitude   = wave_height / 2

                if amplitude > (.001 * ref):  # oscillations must be greater amplitude than .1% of reference value
                    return 'osc'  # exit loop if oscillatory 
        
        ### If there are no oscillations, determine convergence time
        #Reference value for which we compare the previous values to
        ref = state_sol[-1]
        ss.append(ref)
        
        tol = max(ref * .01, 1e-10) # Define threshold at 1% of ref, while addressing exponential decay terms
    
        ### If reference is not zero, work backwards to find time of convergence
        index = 0
        l = len(state_sol) #Amount of time steps in a solution
            
        for i in range(2, l): # Work backwards comparing each time step to the ref
            if not math.isclose(state_sol[-i], ref, abs_tol=tol): 
                index = l - i + 1 #(plus 1 for previous time step, ie. last step that passed the threshold)
                break
        
        conv_index.append(index)
        
    return ss, conv_index
```

### Concentration to individual conversion

```python
def est_viral_abund(r, P, alpha, beta, S_ind, V_ind):
    
    """
    Estimating viral abundance based off ratio of host to viral entitiy.
    
    r     (int)  : ratio of phytoplankton to viral enities, h_ind:v_ind
    P     (float): phytoplankton abundance
    alpha (float): conversion term for host related variables
    beta  (float): conversion term for virus related variables
    S_ind (float): Nucleotides per individual host
    V_ind (float): Nucleotides per individual virus
    """
    if r == 0: 
        return 0.
    
    S_i = P / alpha / S_ind  #individual hosts per ml
 
    V_est = (1/r) * S_i * V_ind * beta #mmol N / m^3

    return V_est
```

```python
def est_viral_ratio(V, P, alpha, beta, S_ind, V_ind): 
    
    """
    Estimating the ratio of viruses to hosts given virus and host concentration
    
    V     (float): viral abundance 
    P     (float): phytoplankton abundance
    alpha (float): conversion term for host related variables
    beta  (float): conversion term for virus related variables
    S_ind (float): Nucleotides per individual host
    V_ind (float): Nucleotides per individual virus
    """
    
    S_i = P / alpha / S_ind  #h_ind / per ml
    
    r = S_i * V_ind * beta / V #h_ind / v_ind 
    
    return r
```
