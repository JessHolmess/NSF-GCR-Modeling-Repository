## Log-Log Plot: Late Time Behavior (SS or Osc?)

```python
%%time

fig, axs = plt.subplots(2, 3, figsize=(15, 7))
axs = axs.reshape(-1)

state_labels = ["N_r", "P_U", "P_I", "Z ", "V_I", "V_E"]

for i in range(6):
    
    for sol in sols:

        z0, state_sol = sol[0], sol[1].y[i+1]
        ref = state_sol[-1]
        
        chunk = int(len(sol[1].t) * .70) # Last 30% of time span
        y_end = state_sol[chunk: ]       # look at tail end of specific state
        
        l_min = argrelmin(y_end)[-1]  # Extracting local minimum 
        l_max = argrelmax(y_end)[-1]  # and maximum
        
        # If there are oscillations....
        if len(l_min == 0) and len(l_max == 0):
            
            # Center point between peak and trough
            ref = (state_sol[chunk + l_min[-1]] + state_sol[chunk + l_max[-1]]) / 2
            
            wave_height = y_end[l_max[-1]] - y_end[l_min[-1]]
            amplitude   = wave_height / 2
            
            if amplitude > (.001 * ref):
                
                # Plot the peak and troughs
                axs[i].scatter(sum(z0), state_sol[chunk + l_min[-1]], color='red', s=1)
                axs[i].scatter(sum(z0), state_sol[chunk + l_max[-1]], color='red', s=1)
            
            else: axs[i].scatter(sum(z0), ref, color='black', s=1)
            
        else: axs[i].scatter(sum(z0), ref, color='black', s=1)
            
    axs[i].set_title(fr"Late Time Behavior of ${state_labels[i]}$")
    
for ax in axs:
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.grid(which='both', alpha=.3)
    ax.set_xlabel(r"Total N (mmol m$^{-3}$)")
    ax.set_ylabel(r"(mmol m$^{-3}$)")

axs[1].set_ylim(1e-12, 1)
axs[2].set_ylim(1e-12, 10)
axs[3].set_ylim(1e-12, 10)
axs[4].set_ylim(1e-12, 1e-1)
axs[5].set_ylim(1e-12, 1e-1)

plt.tight_layout()
plt.show()
```
