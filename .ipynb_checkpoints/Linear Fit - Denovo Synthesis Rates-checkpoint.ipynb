```r
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```r
#Defining data set

irradiance_level = c(25, 150, 300, 500, 1000, 2000)
synthesis_rate = c(1.9, .41, .26, 1, .9, 2)

#The first row could be considered an outlier, as syntesis rate 
#is known to accelerate at higher irradiance levels 
df = data.frame(irradiance_level, synthesis_rate)
summary(df)
```

```r
model = lm(synthesis_rate ~ irradiance_level, data=df)

df = df[-1, ]
no_outlier = lm(synthesis_rate ~ irradiance_level, data=df)
```

```r
summary(model)
```

The linear model for the full data set is weak. Neither of our coefficients are significantly different from zero, the coefficient of determination is very low, and our F-test is insignificant. There is however, on outlier in the data that has a large effect on the model considering how little data we have. 

```r
summary(no_outlier)
```

Running the same linear regression on the data set ecluding the outlier, we get a much better result. We now have one significant coefficient, a large $R^2$, and a significant F-test result. All illustrating this model may be better at representing what little data we have.

```r
plot(irradiance_level, synthesis_rate, pch = 16, cex = 1.3, col = "blue", main = "Irradiance Level vs. Denovo Synthesis Rate", xlab = "Irradiance Level (micromole photons / m^2 s)", ylab = "De novo NT synthesis Rate (day^-1)")
abline(model)
abline(no_outlier, col="red")
text(1400, 1.1, "Excluding outlier", col="red")
text(900, 1.4, "Including outlier")
```
