The [Global Viral Oceanography Database (gVOD)](https://doi.pangaea.de/10.1594/PANGAEA.915758) has been extremely helpful in constraining the parameterization of key viral transport rates. The gVOD is a culmination of over 10,000 measurements of viral abundance and over 700 measurements of viral production metrics paired with relevant oceanographic parameters, pulled from 103 different literature references. You will need to download the csv file of data before running this analysis. 

We analyzed four main viral production metrics, namely virus decay rates, virus-to-prokaryote ratio, burst size, and lysis rate below.

## Plotting Burst Size

```python
# Calculate the inner quartile range
Q3 = df['Burst size [#]'].describe()['75%']
Q1 = df['Burst size [#]'].describe()['25%']

IQR = Q3 - Q1

# If we subtract 1.5 x IQR from the first quartile, any data values that are smaller than this number are considered outliers.
print(f"Burst sizes below {Q1 - (1.5 * IQR)} are considered outliers.")

# If we add 1.5 x IQR to the third quartile, any data values that are greater than this number are considered outliers.
print(f"Burst sizes above {Q3 + (1.5 * IQR)} are considered outliers.")
```

```python
plt.hist(df['Burst size [#]'][df['Burst size [#]'] < 56.5], bins=30, color='black');
plt.hist(df['Burst size [#]'][df['Burst size [#]'] >= 56.5], bins=60, color='red', label='Outliers');

plt.title('Burst Size - Histogram')
plt.xlabel('Burst Size [# VLP / Host]')
plt.ylabel('Measurement Frequency')

plt.figtext(0.665, 0.53, df['Burst size [#]'].describe().to_string())
plt.grid(alpha=.3)
plt.legend()
plt.show()
```

We want to convert viral decay rates from units per hour to units per day. Let $P$ be equal to the proportion of loss per hour. Notice, using a simple exponential decay formula $y(t) = e^{-t / \tau}$, 

$$ (1 - P) = y(1) = e^{-1/\tau} \quad \Rightarrow \quad ln(1-P) = -1 / \tau \quad \Rightarrow \quad \tau = -1 / ln(1-P).$$

Plugging back into the exponential formula gives, 

$$ y(t) = e^{-t / \tau} = e^{t \cdot ln(1-P)} = (e^{ln(1-P)})^{t}) = (1-P)^t \textrm{ for t in hours.}$$

Thus, $y(24) = (1-P)^{24}$ gives the proportion *left over* after 1 day, and $1 - y(24) = 1 - (1-P)^{24}$ is the proportion lost over 1 day. 


Example: if $P = .3$ / hr, then conv_P (converted P) is equal to 

$$\text{conv_P} = 1 - (1 - .3)^{24} = 1 - (.7)^{24} \approx 1 - (0.0001915) \approx 0.9998 \text{ per day}.$$

## Calculating Viral Decay Rates: assuming exponential decay rate constant reported

```python
# Simple conversion: 24 * per hour = per day
viral_decay_rate_per_day = 24 * (viral_decay_perc_per_hour / 100)
```

```python
# Calculate the inner quartile range
Q3 = viral_decay_rate_per_day.describe()['75%']
Q1 = viral_decay_rate_per_day.describe()['25%']

IQR = Q3 - Q1

# If we subtract 1.5 x IQR from the first quartile, any data values that are smaller than this number are considered outliers.
print(f"Viral Decay Rates below {Q1 - (1.5 * IQR)} are considered outliers.")

# If we add 1.5 x IQR to the third quartile, any data values that are greater than this number are considered outliers.
print(f"Viral Decay Rates above {Q3 + (1.5 * IQR)} are considered outliers.")
```

```python
plt.hist(viral_decay_rate_per_day[viral_decay_rate_per_day < 1.0644], bins=40, color='black');
plt.hist(viral_decay_rate_per_day[viral_decay_rate_per_day >= 1.0644], bins=150, color='red', label = "Outliers");

plt.title('Viral Decay Rate / Day - Histogram')
plt.xlabel('Proportion of Decay per Day')
plt.ylabel('Measurement Frequency')
plt.xlim(0, 1.5)

plt.figtext(0.67, 0.53, viral_decay_rate_per_day.describe().to_string())
plt.grid(which='both', alpha=.3)
plt.legend()
plt.show()
```

## Calculating Lysis Rates: assuming exponential decay rate constant reported

```python
lysis_perc_per_hour = df['Cells lysed [%/h]'][df['Cells lysed [%/h]'] != '#999.000000000'].dropna().astype(float)
lysis_rate_per_day = 24 * (lysis_perc_per_hour / 100)
```

```python
# Calculate the inner quartile range
Q3 = lysis_rate_per_day.describe()['75%']
Q1 = lysis_rate_per_day.describe()['25%']

IQR = Q3 - Q1

# If we subtract 1.5 x IQR from the first quartile, any data values that are smaller than this number are considered outliers.
print(f"Lysis Rates below {Q1 - (1.5 * IQR)} are considered outliers.")

# If we add 1.5 x IQR to the third quartile, any data values that are greater than this number are considered outliers.
print(f"Lysis Rates above {Q3 + (1.5 * IQR)} are considered outliers.")
```

```python
plt.hist(lysis_rate_per_day[lysis_rate_per_day < 0.775], bins=20, color='black');
plt.hist(lysis_rate_per_day[lysis_rate_per_day >= 0.775], bins=50, color='red', label = "Outliers");

plt.title('Lysis Rate / Day - Histogram')
plt.xlabel('Proportion of Hosts Lysed per Day')
plt.ylabel('Measurement Frequency')

plt.figtext(0.67, 0.53, lysis_rate_per_day.describe().to_string())
plt.grid(which='both', alpha=.3)
plt.legend()
plt.show()
```
