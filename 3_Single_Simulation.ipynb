## Define Initial State Conditions

```python
'''
Defining initial state conditions for numerical integration.
Parameters units of mmol m^-3, must be type float (for printing purposes). 
'''

# Define total N of the system. Can be almost any value, 
# but system dynmaics are interesting between .003 and 20 mmol m^-3
Nt_0 = .025

# Define initial phytoplankton and zoolplankton 
# concentrations as percentages of given total N
Pu_0 = .1 * Nt_0  
Z_0  = .2 * Nt_0

# Alternate nutrients, infected phyto, and intra virus
# can be initialized at 0% of total N
Nn_0 = 0.   
Pi_0 = 0.   
Vi_0 = 0. 

# Initialize free viral abundance based off
# initial phytoplankton concentration
r = 1/10 # ratio of hosts:free-virus 
Ve_0 = est_viral_abund(r, Pu_0, alpha, beta, S_ind, V_ind)

# Define nutrient pool as left over concentration
Nr_0 = Nt_0 - (Pu_0 + Z_0 + Ve_0)

# Define initial condition array to pass to numerical integrator
z0 = [Nn_0, Nr_0, Pu_0, Pi_0, Z_0, Vi_0, Ve_0]

# Check total N is at level intended
print(f"Total N: {sum(z0)}")
```

## Nicely Print Initial Conditions and Parameterization

```python
initial_names = ['Nn_0', 'Nr_0', 'Pu_0', 'Pi_0', 'Z_0', 'Vi_0', 'Ve_0']
print()

print(f'N_t = {sum(z0):.4}', end='  ')
for i, state in enumerate(initial_names):
    print(f'{state} = {z0[i]:.5}', end='  ')
print(end='\n \n')

print_parameterization(param)
```

## Run Numerical Integrator

```python
%time 

# Defining length of time span
t = 500 # days

# Maximum time step length (days) allowed
max_step = .1

# Integration Method Used 
method = 'LSODA'

# Definition individual absolute tolerances for each state variable
# always defined in the same order ['Nn', 'Nr', 'Pu', 'Pi', 'Z', 'Vi', 'Ve']
tolerances = [1e-8, 1e-8, 1e-12, 1e-8, 1e-8, 1e-10, 1e-10]

# Running numerical integrator
sol = solve_ivp(model, [0, t], z0, args=param, max_step=max_step, method=method, atol=tolerances)
```

## Visualizing the Solutions

```python
# Defining each state array from solve_ivp solution
N_n = sol.y[0] # New Nutrients
N_r = sol.y[1] # Recycled Nutrients
P_U = sol.y[2] # Susceptible Hosts
P_I = sol.y[3] # Infected Hosts
Z   = sol.y[4] # Zooplankton
V_I = sol.y[5] # Internal viruses 
V_E = sol.y[6] # Free (extracellular) viruses 

# Initializing figure
fig = plt.figure(figsize = (12, 7))
ax = fig.subplot_mosaic("""A;A;A;B;B;B;C;C""")

for label in ['A', 'B']:
    ax[label].plot(sol.t, P_U, color="mediumaquamarine", label='Uninfected Phytoplankton')
    ax[label].plot(sol.t, P_I, color="yellowgreen"     , label='Infected Phytoplankton')
    ax[label].plot(sol.t, Z,   color="darkgreen"       , label='Zooplanton')
    ax[label].plot(sol.t, V_I, color="red"             , label='Intracellular Virus')
    ax[label].plot(sol.t, V_E, color="darkred"         , label='Extracellular Virus')
    ax[label].plot(sol.t, N_n, color="darkviolet"      , label='Nitrate')
    ax[label].plot(sol.t, N_r, color="blue"            , label='Recycled Nutrients')

# Linear Plot Settings
title = (f'Total N: {Nt_0 :.8}, '
         f'for Nr_0 = {Nr_0:.4}, '
         f'Pu_0 = {Pu_0:.4}, '
         f'Pi_0 = {Pi_0:.4}, '
         f'Z_0 = {Z_0:.4}, '
         f'Vi_0 = {Vi_0:.4}, '
         f'Ve_0 = {Ve_0:.4}\n\n'
         'System Dynamics - Linear Scale')

ax["A"].set_title(title, pad=15)

# Log Plot Settings
ax["B"].set_title('System Dynamics - Log Scale')
ax["B"].set_yscale('log')
ax["B"].set_ylim(1e-12, 1)

# Dotted line depicting damping and adsorptino cut-off 
ax["B"].axhline(y=1e-12, xmin=-20, xmax=600, c='grey', linestyle='--', alpha=.5)

# Sum of states compared to prescribed total N - should be consistent over time
ax["C"].plot(sol.t, N_n + N_r + P_U + P_I + Z + V_I + V_E, color ='darkorange', label='Numerical Total Sum')
ax["C"].plot(sol.t, np.full(len(sol.t), sum(z0)), '--', color = 'red', label='True Total N Sum')
ax["C"].set_title('Total Nutrients')

# General Plot Specificities
for plot in ax:
    ax[plot].set_xlabel('Time (Day)')
    ax[plot].set_ylabel(r'Nutrients ($mmol/m^3$)')
    
    ax[plot].grid(alpha=.3)
    ax[plot].legend(loc=(1.01, 0), framealpha=1)

plt.tight_layout()
plt.show()
```
