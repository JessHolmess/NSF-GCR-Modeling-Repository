```python
import numpy as np
import matplotlib.pyplot as plt
from math import isclose 

#defining functional form for Pi

def steady_Pi(V_I, alpha, beta, V_max, γ_Z, g, λ_P, λ_Z, λ_Z_hat, δ, μ_V, K_P):
    
    """Pos and Neg represent positive and negative branches of the quadratic solution"""
    
    ratio = λ_Z_hat * K_P ** 2 / γ_Z * g ** 2
    
    R = V_I - (K_P * λ_Z / γ_Z * g) - ratio * (V_max - λ_P)
    M = ratio * (μ_V + δ)
    
    P_I_pos = - (R + alpha * V_I / beta) / 2 + np.sqrt((R + alpha * V_I / beta) ** 2 
                                                       - (4 * alpha * V_I * (R + M)) / beta) / 2
    
    P_I_neg = - (R + alpha * V_I / beta) / 2 - np.sqrt((R + alpha * V_I / beta) ** 2 
                                                       - (4 * alpha * V_I * (R + M)) / beta) / 2
    
    return P_I_pos, P_I_neg
```

```python
alpha = (1 / 1.42e8) * 2.1e-10 * 1e6 # host conversion, mmol ml / NT m^3
beta  = (1 / 4e5) * 1.27e-15 * 1e6   # viral conversion, mmol ml / NT m^3

V_max   = 1.4   # Host growth rate
γ_Z     = .4    # Growth rate / Assimilation frequency 
g       = 1.4   # d^-1 Maximum specific grazing rate
λ_P     = .05   # Mortality & Respiration rate
λ_Z     = .05   # d^-1, Linear mortality & respiration & egestion
λ_Z_hat = .1    # Quadratic mortality & respiration & egestion
δ       = .45   # Host lysis rate
μ_V     = .5    # Max host NT recycling rate 
K_P     = 2.8   # mmol m^-3, Half-saturation constant for grazing of P

V_I     = np.linspace(0, 6, 10000)

P_I_pos, P_I_neg = steady_Pi(V_I, alpha, beta, V_max, γ_Z, g, λ_P, λ_Z, λ_Z_hat, δ, μ_V, K_P)
```

```python
plt.figure(figsize=(6,5))

plt.plot(V_I[np.where(P_I_pos >= 0)], P_I_pos[np.where(P_I_pos >= 0)], 
         color='blue', label=r"+ Branch, + P_I")
plt.plot(V_I[np.where(P_I_pos < 0)], P_I_pos[np.where(P_I_pos < 0)], '--', 
         color='blue', label=r"+ Branch, $-$ P_I")

plt.plot(V_I, P_I_neg, '--', color='red', label=r"$-$ Branch, $-$ P_I")

plt.title("Quadratic SS Sol for P_I, derived from NrPiZViVe")
plt.xlabel(r"V_I (mmol N / $m^3$)")
plt.ylabel(r"P_I (mmol N / $m^3$)")

plt.xlim(1e-5, 1e2)
plt.ylim(-1, 6)
plt.xscale("log")

plt.grid(alpha=.3)
plt.margins(2)
plt.legend()
```
